7. Outliers Handling

```python
# Calculate IQR
Q1 = df_dataco["delivery_delay"].quantile(0.25)
Q3 = df_dataco["delivery_delay"].quantile(0.75)
IQR = Q3 - Q1

# Define Outlier Range
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Extract Outliers
outliers = df_dataco[
    (df_dataco["delivery_delay"] < lower_bound) |
    (df_dataco["delivery_delay"] > upper_bound)
]

# View Results
outliers.shape
outliers.head()

```

8. Feature Engineering

```python
df_dataco["delivery_delay"]=df_dataco["days_for_shipping_(real)"]-df_dataco["days_for_shipment_(scheduled)"]
```
